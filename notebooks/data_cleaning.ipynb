```python
#cleaning the existing csv files using pandas

import pandas as pd

# Veri setlerini yükle
# Not: Kullanıcı 'exercises.csv.xlsx - exercises.csv' dosyasını yüklemişti.
df_kaggle = pd.read_excel("data/database/exercises.csv.xlsx")

# 1. Instructions (Talimatlar) sütunlarını birleştirme
instruction_cols = [col for col in df_kaggle.columns if 'instructions' in col]
# Her satır için dolu olan instruction sütunlarını birleştiriyoruz.
df_kaggle['description'] = df_kaggle[instruction_cols].apply(lambda x: ' '.join(x.dropna().astype(str)), axis=1)

# 2. Secondary Muscles (İkincil Kaslar) sütunlarını birleştirme
secondary_cols = [col for col in df_kaggle.columns if 'secondaryMuscles' in col]
df_kaggle['secondary_muscles_combined'] = df_kaggle[secondary_cols].apply(lambda x: ', '.join(x.dropna().astype(str)), axis=1)

# 3. Yeni DataFrame oluşturma (Kullanıcının şablonuna uygun)
new_df = pd.DataFrame()

new_df['id'] = df_kaggle['id']
new_df['name'] = df_kaggle['name']
# Kullanıcının isteği üzerine 'target' sütununu 'primary_muscles' olarak alıyoruz.
new_df['primary_muscles'] = df_kaggle['target'] 
new_df['secondary_muscles'] = df_kaggle['secondary_muscles_combined']
new_df['body_part'] = df_kaggle['bodyPart'] # Filtreleme için yararlı olabilir diye tutuyoruz
new_df['equipment'] = df_kaggle['equipment']
new_df['difficulty'] = None # Doldurulacak
new_df['risk_level'] = None # Doldurulacak
new_df['impact'] = None # Doldurulacak
new_df['suitable_for'] = None # Doldurulacak
new_df['!suitable_for'] = None # Doldurulacak (Kritik)
new_df['description'] = df_kaggle['description']
#new_df['gif_url'] = df_kaggle['gifUrl']

# 4. CSV olarak kaydetme
output_filename = 'exercises_cleaned.csv'
new_df.to_csv(output_filename, index=False)

print(f"Dosya '{output_filename}' adıyla oluşturuldu.")
print(new_df.head().to_markdown(index=False, numalign="left", stralign="left"))
```

```python
#deciding on the difficulty levels based on exercise names and equipment
import pandas as pd

df = pd.read_csv("data/database/exercises_by_muscle_n_eq.csv")

def refine_difficulty(row):
    name = row['name'].lower()
    equipment = row['equipment']
    
    # --- 1. Adım: Kelime Bazlı "Zor" Hareketler ---
    # Bu kelimeler geçiyorsa kesinlikle ZORDUR (Advanced)
    hard_keywords = ['pull-up', 'chin-up', 'muscle-up', 'dip', 'dips', 'handstand', 'pistol', 'clean', 'snatch']
    if any(word in name for word in hard_keywords):
        return 'Advanced'

    # --- 2. Adım: Kelime Bazlı "Kolay" Hareketler ---
    # Bu kelimeler geçiyorsa genelde KOLAYDIR (Beginner)
    easy_keywords = ['stretch', 'wall', 'kneeling', 'march', 'assisted']
    if any(word in name for word in easy_keywords):
        return 'Beginner'
    
    beginner_keywords = ['leverage machine', 'band', 'cable', 'resistance band', 'roller', 'sled machine', 'assisted', 'medicine ball']
    intermediate_keywords = ['dumbbell', 'barbell', 'ez barbell', 'kettlebell']


    # --- 3. Adım: Geriye Kalanlar İçin Ekipman Kontrolü ---
    if equipment in beginner_keywords:
        return 'Beginner'
    elif equipment in intermediate_keywords:
        return 'Intermediate' # Barbell ama clean/snatch değilse ortadır
    elif equipment == 'bodyweight':
        return 'Intermediate' # Barfiks değilse, squat/lunge gibi ortadır
    else:
        return 'Intermediate'

# Fonksiyonu Uygula
df['difficulty'] = df.apply(refine_difficulty, axis=1)

output_filename = 'exercises_dif0.csv'
df.to_csv(output_filename, index=False)
```

```python
import json

def add_safety_tag(input_file, output_file):
    # 1. Load the existing JSON
    with open(input_file, 'r', encoding='utf-8') as f:
        data = json.load(f)

    # 2. Iterate and add the new tag
    for exercise in data:
        # We initialize it as an empty list
        # This prevents errors in your RAG pipeline later
        if "not_suitable_for" not in exercise:
            exercise["not_suitable_for"] = []

    # 3. Save the modified JSON
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(data, f, indent=4)
    
    print(f"Structure updated! Saved to {output_file}")

# Run the script
add_safety_tag('exercises_dif0.json', 'exercises1.json')
```

```python
import json

RULES = {
    "lowerback_pain": ["deadlift", "back squat", "sit-up", "spinal loading", "crunch", "hyperextension", "handstand", "side plank hip adduction", "spine", "good morning","barbell row", "snatch"],
    "knee_pain": ["jump", "lunge", "running", "sprint", "box jump", "jump squat", "run", "mountain climbers", "plyo", "pistol", "squat", "snatch"],
    "shoulder_pain": ["overhead press", "behind the neck", "upright row", "snatch", "body saw", "pull-up", "chin-up", "dips", "dip", "handstand", "traps", "delts", "overhead", "snatch", "lateral"],
    "hypertension": ["heavy", "isometric", "sprint", "breath hold", "weighted crunch", "hanging leg raise", "decline", "snatch"],
    "neck_pain": ["overhead press", "shoulder press", "bridge", "shrug", "handstand", "levator scapulae", "deadlift", "rack pull", "snatch", "behind head"],
    "obesity": ["pull-up", "chin-up", "hanging", "run", "stride", "sprint", "jump", "burpee", "plyo", "pistol", "handstand", "side plank hip adduction"]
}

# These keywords will PREVENT the "lowerback_pain" tag from being added
# because they are actually good for rehab.
SAFE_CORE_KEYWORDS = ["plank", "dead bug", "bird dog", "pelvic tilt", "glute bridge"]
SAFE_FOR_ALL = ["stretch", "resistance band"]

muscle_list = ["abductors", "abs", "adductors", "biceps", "calves", "delts", "cardiovascular system", "forearms", "glutes", "hamstrings", "lats", "pectorals", "levator scapulae", "quads", "serratus anterior", "spine", "traps", "triceps", "upper back"]

def bulk_tag_exercises(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as f:
        data = json.load(f)

    changed_count = 0

    for ex in data:
        # 1. Start with existing tags
        tags = set(ex.get("not_suitable_for", []))
        
        # 2. Extract muscles and handle both String and List formats
        raw_muscles = ex.get('primary_muscles', [])
        if isinstance(raw_muscles, str):
            ex_muscles = [raw_muscles.lower()]
        else:
            ex_muscles = [m.lower() for m in raw_muscles]

        # 3. Use membership check
        # We check if any of the exercise muscles overlap with our target list
        if any(m in muscle_list for m in ex_muscles):
            content = (ex['name'] + " " + (ex.get('description') or "")).lower()
            
            # 4. Apply the keyword rules
            for condition, keywords in RULES.items():
                if any(word in content for word in keywords):
                    # SAFETY LOGIC: Don't tag rehab exercises as bad for back pain
                    if condition == "lowerback_pain" and any(safe in content for safe in SAFE_CORE_KEYWORDS):
                        continue
                    if any(safe in content for safe in SAFE_FOR_ALL) or ((ex['equipment'] == 'band' or ex['equipment'] == 'body weight' or ex['equipment'] == 'assisted') and ex['difficulty'] == 'Beginner'):
                        continue
                    tags.add(condition)
        
        # 5. Only update if something changed
        new_tags_list = list(tags)
        if len(new_tags_list) > len(ex.get("not_suitable_for", [])):
            changed_count += 1
            
        ex["not_suitable_for"] = new_tags_list

    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(data, f, indent=4)
    
    print(f"Bulk tagging complete! Updated {changed_count} exercises.")

bulk_tag_exercises('exercises3.json', 'exercises4.json')
```

```python
import json

# Load data
with open('data/database/exercises_simple.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# Keywords
PUSH_KEYWORDS = ["push", "pushdown", "press", "triceps", "crusher", "dip"]
PULL_KEYWORDS = ["row", "pull", "pulldown", "pull-up", "chin-up", "hyperextension", "shrug"]
HINGE_KEYWORDS = ["hinge", "deadlift", "bridge", "good morning", "rack pull"]
SQUAT_KEYWORDS = ["squat", "step-up", "lunge", "leg press"]
EXTENSION_KEYWORDS = ["leg curl", "leg extension", "thrust", "hip extension", "glute kickback", "abduction", "adduction"]

for ex in data:
    # 1. Normalize for comparison
    name = ex.get('name', '').lower()
    body_part = ex.get('body_part', '').lower()
    equipment = ex.get('equipment', '').lower()
    primary = [m.lower() for m in ex.get('primary_muscles', [])]

    # 2. Equipment Cleanup
    if equipment == "leverage machine":
        ex['equipment'] = "machine"
    elif equipment == "resistance band":
        ex['equipment'] = "band"

    # 3. Categorize TYPE (Mobility, Cardio, then Strength)
    if "stretch" in name:
        ex['type'] = 'mobility'
    elif body_part == "cardio":
        ex['type'] = "cardio"
    else:
        ex['type'] = "strength"

    # 4. Categorize PATTERN (Only for strength)
    if ex['type'] == "strength":
        # Handle Legs
        if body_part == "legs":
            if any(k in name for k in SQUAT_KEYWORDS):
                ex['pattern'] = "squat"
            elif any(k in name for k in HINGE_KEYWORDS):
                ex['pattern'] = "hinge"
            elif any(k in name for k in EXTENSION_KEYWORDS):
                ex['pattern'] = "extension"
        
        # Handle Upper Body
        elif body_part in ['chest', 'shoulders'] or "triceps" in primary:
            if any(k in name for k in PUSH_KEYWORDS) or "triceps" in primary:
                ex['pattern'] = "push"
        
        elif body_part == 'back' or "biceps" in primary:
            if any(k in name for k in PULL_KEYWORDS) or "biceps" in primary:
                ex['pattern'] = "pull"
        
        elif body_part == 'core':
            ex['pattern'] = "core"
            
    # Fallback for anything not caught
    if 'pattern' not in ex:
        ex['pattern'] = ''

# Save the updated data
with open('data/database/exercises_s2.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, indent=4)
```
